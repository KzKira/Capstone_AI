```python
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
import glob
```

```python
IMAGE_SHAPE = (224,224,3)
```

```python
def decode_label(label: np.ndarray, img_shape: tuple):
    label_shape = label.shape

    label = np.concatenate([label[...,1:2], label[...,0:1], label[...,2:]], axis=-1) # index: ij

    grid_coord = np.stack(np.mgrid[:label_shape[0], :label_shape[1]], axis=-1)
    pad = np.zeros((*label_shape[:2], label_shape[-1]-2))
    grid_coord = np.concatenate([grid_coord,pad], axis=-1)
    grid_coord = np.expand_dims(grid_coord, axis=-2)
    grid_coord = np.repeat(grid_coord, label_shape[2], -2)

    label = label + grid_coord # 그리드 좌표를 추가

    grid_height = img_shape[0] / label_shape[0]
    grid_width = img_shape[1] / label_shape[1]

    rewinder = np.array([grid_height, grid_width, img_shape[1], img_shape[0], 1, 1, 1], np.float32) # y,x,width,height

    label = label * rewinder # 다시 이미지 기준으로 돌리기

    bboxes = np.round(label[label[...,4] == 1.0]) # Confidence가 1이면 bounding box
    bboxes = np.concatenate([bboxes[...,1:2], bboxes[...,0:1], bboxes[...,2:]], axis=-1) # index: xy
    return bboxes
```

```python
def parse_tfrecord_fn(example):
    feature_description = {
        'image': tf.io.FixedLenFeature((), tf.string),
        'label': tf.io.FixedLenFeature((16,16,6,7), tf.float32),
    }
    
    example = tf.io.parse_single_example(example, feature_description)

    image = tf.io.decode_jpeg(example["image"], channels=3)
    image = tf.reshape(image, (224,224,3))

    label = example["label"]
    label = tf.reshape(label, (16,16,6,7))
    
    return image, label


train_records = sorted(glob.glob("./train/*.tfrecord"))

raw_dataset = tf.data.TFRecordDataset(train_records, compression_type="GZIP")
images = raw_dataset.map(parse_tfrecord_fn).batch(16)
it = images.as_numpy_iterator()
```

```python
plt.figure(figsize=(11,10))
images, labels = next(it)
for idx, (image, label) in enumerate(zip(images, labels)):
    inp = np.array(image, np.uint8)

    for bbox in decode_label(label, IMAGE_SHAPE):
        x, y, w, h = bbox[:4]

        left = int(x-w/2)
        right = int(x+w/2)
        top = int(y-h/2)
        bottom = int(y+h/2)

        cv2.rectangle(inp, (left, top), (right, bottom), (0,255,0), 3)

    plt.subplot(4,4,idx+1)
    plt.imshow(inp)
```
